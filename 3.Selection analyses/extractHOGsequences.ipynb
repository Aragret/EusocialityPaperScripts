```python
import pandas as pd
import os
from Bio import SeqIO
import matplotlib.pyplot as plt
```

```python
# For single-copy orthologs

HOGs_table = pd.read_table("/EBB-ZIVNFS/amikhail/termite_genomes/eusociality_paper/proteomes/OrthoFinder/Results_Dec13_1/Phylogenetic_Hierarchical_Orthogroups/SingleCopyHOGs_dir/SingleCopyHOGs",
                          sep = "\t")
HOGs_table.head()
```

```python
HOGs_table.tail()
```

```python
# check for empty values
HOGs_table.isnull().sum().sum()
```

```python
HOGs_table = HOGs_table.dropna()
```

```python
os.chdir("/EBB-ZIVNFS/amikhail/termite_genomes/eusociality_paper/proteomes/OrthoFinder/")
```

```python
HOGs_table.OG
```

```python
for og in HOGs_table.OG:
    single_HOGs = []
    temp = HOGs_table[HOGs_table["OG"] == og]
    HOG_name = temp.HOG
    column_number = len(temp.columns)
    with open("Results_Dec12_1/Orthogroup_Sequences/{0}.fa".format(og)) as seqs:
        for record in SeqIO.parse(seqs, "fasta"):
            # check if gene (record.id) is present in the single-copy HOGs table
            mask = temp.isin([record.id])
            if temp[mask].isnull().sum().sum() == (column_number - 1):
                single_HOGs.append(record)
    SeqIO.write(single_HOGs, "Results_Dec13_1/Phylogenetic_Hierarchical_Orthogroups/SingleCopyHOGs_dir/{0}.fa".format(og), "fasta")
    print(len(single_HOGs))
```
