```python
### parsing RELAX bootstrap

# for i in range(1, 101):
for i in range(35, 101):
    print(i)
    os.chdir('/EBB-ZIVNFS/amikhail/termite_genomes/eusociality_paper/analyses/relax/bootstrap_{0}'.format(i))
    with open("/EBB-ZIVNFS/amikhail/termite_genomes/eusociality_paper/analyses/relax/bootstrap_{0}/relax_table.tsv".format(i), "w") as out:
        directory = os.fsencode("/EBB-ZIVNFS/amikhail/termite_genomes/eusociality_paper/analyses/relax/bootstrap_{0}/".format(i))
        out.write("HOG\tLRT\tp-value\tk\n")
        for file in os.listdir(directory):
            filename = os.fsdecode(file)
            if filename.endswith(".Output") and os.stat(filename).st_size != 0:
                HOG = filename.split("-")[0]
                print(filename)
                with open(filename) as f:
                    data = f.read()
                    new_data = re.sub(":inf|:-nan", ":0", data)
                    output_dict = json.loads(new_data)
                    values = list(output_dict["test results"].values())
                    out.write("{0}\t{1}\t{2}\t{3}\n".format(HOG, values[0], values[1], values[2]))
```

```python
df = pd.DataFrame(columns=['HOG', 'LRT', 'p-value', 'k'])
for i in range(100):
    try:
        df_branch = pd.read_table('/EBB-ZIVNFS/amikhail/termite_genomes/eusociality_paper/analyses/relax/bootstrap_{0}/relax_table.tsv'.format(i + 1))
        # print(df.head())
        df_branch['Bootstrap'] = 'bootstrap_{0}'.format(i + 1)
        df = pd.concat([df, df_branch])
    except FileNotFoundError:
        print('branch_{0}'.format(i + 1))
        continue
```
