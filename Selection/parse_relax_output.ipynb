```python
# change working_dir to your directory with RELAX output
# HOG - orthogroup name

# os.chdir('/EBB-ZIVNFS/amikhail/termite_genomes/eusociality_paper/analyses/relax/all_internal')
with open("/working_dir/relax_table.tsv", "w") as out:
    directory = os.fsencode("/working_dir/")
    out.write("HOG\tLRT\tp-value\tk\n")
    for file in os.listdir(directory):
        filename = os.fsdecode(file)
        if filename.endswith(".Output") and os.stat(filename).st_size != 0:
            
            # change to get the orthogroup name from the filename if needed
            HOG = filename.split("-")[0]
            
            with open(filename) as f:
                data = f.read()
                output_dict = json.loads(data)
                values = list(output_dict["test results"].values())
                out.write("{0}\t{1}\t{2}\t{3}\n".format(HOG, values[0], values[1], values[2]))
```
